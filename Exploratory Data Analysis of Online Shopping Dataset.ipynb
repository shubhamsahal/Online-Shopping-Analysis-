# 2) DATA VISUALIZATION

```python
#  tenure_month of the customer

order = shop["Tenure_Months"].value_counts().iloc[:20].index

plt.figure(figsize=(12, 6))       
sns.countplot(x="Tenure_Months", data=shop, order=order, palette="hls")
plt.title("Customer and their tenure_Month")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

# Many customers typically stay around 40 months on the platform for their purchases.





```python
# sales in monthwise

count=shop["Month"].value_counts().iloc[:12].index

plt.figure(figsize=(10,4))
sns.countplot(x = shop["Month"], data = shop, order = count, palette = "viridis")
plt.title("sales in a Month")
plt.xticks(rotation=40, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

# August and June exhibit higher sales, while February experiences comparatively lower sales.

```python
# sales of product based on category

count=shop["Product_Category"].value_counts().iloc[:20].index

plt.figure(figsize=(10,6))
sns.countplot(y=shop["Product_Category"],order=count,data=shop,palette="viridis")
plt.title("sales of product")
plt.xticks(rotation=40, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

# This indicates that product categories such as Apparel, Nest-USA, and Office products are the most frequently sold. 

```python
# Customer based on Gender
gender_values = shop["Gender"].value_counts()

# Pie chart customization
colors = ["#ff6e95", "#6ecfff"]  # Custom colors for Female and Male
explode = [0.1, 0.1]  # Explode Female and Male slices for emphasis

plt.figure(figsize=(8, 8))
plt.pie(gender_values, labels=["Female", "Male"], colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=90)
plt.title("Customer Based on Gender")
plt.legend(title="Gender", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))  # Add legend for clarity
plt.show()
```

# This analysis indicates that there are a higher number of female customers compared to male customers.

```python
# Customer based on Location
location_values = shop["Location"].value_counts()

# Pie chart customization
colors = plt.cm.tab10.colors[:len(location_values)]  # Using a colormap for varied colors
explode = [0.1] * len(location_values)  # Explode slices for emphasis

plt.figure(figsize=(8, 8))
plt.pie(location_values, labels=location_values.index, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=90)
plt.title("Customer Distribution by Location")
plt.legend(title="Location", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))  # Add legend for clarity
plt.show()
```

# This chart indicates that a larger proportion of customers, around 34%, are from Chicago, with a notable presence from California locations.

```python
# GST density plot
sns.set_style("darkgrid")
sns.kdeplot(shop["GST"], color="red", fill=True, label="GST")
plt.xlabel("GST")
plt.ylabel("Density")
plt.title("GST Density Plot")
plt.xticks(rotation=20, ha="right")  # Rotate x-axis labels for better readability
plt.legend()
plt.show()

# Quantity density plot
sns.kdeplot(shop["Quantity"], color="blue", fill=True, label="Quantity")
plt.xlabel("Quantity")
plt.ylabel("Density")
plt.title("Quantity Density Plot")
plt.xticks(rotation=20, ha="right")  # Rotate x-axis labels for better readability
plt.legend()
plt.show()
```

# The GST density plot highlights popular purchase choices with 0.18% and 0.10%, while the Quantity plot shows a preference for smaller product quantities.

# 3) TIME SERIES ANALYSIS


```python
# Month-wise offline and online sales comparison
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Offline sales
shop.groupby("Month")["Offline_Spend"].agg("sum").plot(ax=axes[0], color="blue", marker='o')
axes[0].grid()
axes[0].set_title("Month-wise Offline Sales")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales in Lakh")

# Online sales
shop.groupby("Month")["Online_Spend"].agg("sum").plot(ax=axes[1], color="red", marker='o')
axes[1].grid()
axes[1].set_title("Month-wise Online Sales")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Sales in Lakh")

plt.tight_layout()
plt.show()
```

# Online and offline spending exhibit similar growth patterns, with notable peaks observed in August and December.

```python
# Total product sales in a month
plt.figure(figsize=(10, 4))
colors = plt.cm.viridis(np.linspace(0, 0.8, 12))

shop.groupby("Month")["Product_Category"].agg("count").plot(kind="bar", color=colors)
plt.title("Monthly Total Product Sales")
plt.xlabel("Month")
plt.ylabel("Count of Product Sales")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

plt.show()
```

# The bar chart indicates that the highest number of products were sold in the month of August.

```python
# Product sales in each location
plt.figure(figsize=(10, 4))
colors = plt.cm.viridis(np.linspace(0, 0.9, 5))

shop.groupby("Location")["Product_Category"].agg("count").plot(kind="bar", color=colors)
plt.title("Product Sales in Each Location")
plt.xlabel("Location")
plt.ylabel("Count of Product Sales in a Location")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

plt.show()
```

# The bar chart indicates that the highest number of products were sold in the locations of Chicago (greater than 17,500) and California (greater than 15,000).

```python
# Top 5 categories by Online_Spend
online_category = shop.groupby("Product_Category")["Online_Spend"].sum().reset_index().sort_values(by="Online_Spend", ascending=False).head(5)

plt.figure(figsize=(10, 4))
sns.lineplot(x="Product_Category", y="Online_Spend", data=online_category, marker='o', color='blue')
plt.title("Top 5 Categories by Online Spending")
plt.xlabel("Product Category")
plt.ylabel("Online Spend")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Top 5 categories by Offline_Spend
offline_category = shop.groupby("Product_Category")["Offline_Spend"].sum().reset_index().sort_values(by="Offline_Spend", ascending=False).head(5)

plt.figure(figsize=(10, 4))
sns.lineplot(x="Product_Category", y="Offline_Spend", data=offline_category, marker='o', color='green')
plt.title("Top 5 Categories by Offline Spending")
plt.xlabel("Product Category")
plt.ylabel("Offline Spend")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

# The line plots reveal that both offline and online spending are higher in categories such as Apparel and Nest-USA products.








# 3)  Model building 

```python
# Select only numeric columns
numeric_columns = shop.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix for numeric variables
numeric_correlation_matrix = numeric_columns.corr()

# Plot the heatmap for numeric variables
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```
